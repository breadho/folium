## 1일차 데이터에 대한 구간별 ANOVA 실시

```python
import scipy.stats as stats
```

```python
## 정규성 검정 & 왜도 검사

for index, data_seq in enumerate(data_list):
    print('\n', "-"*50, '\n', f'test{index + 1}', '\n')
    for data in data_seq:
        print( stats.shapiro( data )
              , 'skewness test: {}'.format( abs(stats.skew(data , bias = False) ) ) 
              , sep = '\n'
             )
```

```python
## 등분산성 검정
## Bartlett’s test tests the null hypothesis that all input samples are from populations with equal variances.
## So best case scenario is not to reject null hypothesis.
from scipy.stats import bartlett

for index, data_seq in enumerate(data_list):
    print('\n', "-"*50, '\n', f'test{index + 1}', '\n')
    print(stats.bartlett(data_seq[0], data_seq[1], data_seq[2], data_seq[3], data_seq[4]))
    print(stats.levene(data_seq[0], data_seq[1], data_seq[2], data_seq[3], data_seq[4]))

```

### ANOVA 실시

```python
for index, data_seq in enumerate(data_list):
    print('\n', "-"*100, '\n', f'test{index + 1}', '\n')

    # ANOVA 수행
    statistic, p_value = stats.f_oneway(data_seq[0], data_seq[1], data_seq[2], data_seq[3], data_seq[4])

    # 결과 출력
    print("ANOVA Statistic:", statistic)
    print("p-value:", p_value)

    # p-value를 기준으로 유의수준을 설정하고 결과를 해석
    alpha = 0.05
    if p_value < alpha:
        print("p-value가 유의수준 {} 보다 작으므로 그룹 간에 통계적으로 유의한 차이가 있습니다.".format(alpha))
    else:
        print("p-value가 유의수준 {} 이상이므로 그룹 간에 통계적으로 유의한 차이가 없습니다.".format(alpha))
```

```python
for index, data_seq in enumerate(data_list):
    print('\n', "-"*100, '\n', f'test{index + 1}', '\n')

    # ANOVA 수행
    F, p = stats.f_oneway(data_seq[0], data_seq[1], data_seq[2], data_seq[3], data_seq[4])
    print( 'F-Ratio: {}'.format(F) , 'p-value:{}'.format(p), sep = '\n')
```

## 지도 위에 시각화

```python
import folium
import pandas as pd

# Folium 지도 객체 생성
m = folium.Map(location=[sampled_data['Latitude'].mean(), sampled_data['Longitude'].mean()], zoom_start=10)

# course 값에 따라 다른 색상 부여
color_dict = {'A': '#4C749F', 'B': '#CC8133', 'C': '#5E9241', 'D' : '#A83D3E', 'E' : '#8B72B0'}

# 데이터프레임 순회하면서 지도에 PolyLine으로 경로 표시
for course, group in sampled_data.groupby('course'):
    locations = list(zip(group['Latitude'], group['Longitude']))
    folium.PolyLine(locations=locations, color=color_dict[course], weight=10, opacity=1, popup=f'Course: {course}').add_to(m)

# 지도를 HTML 파일로 저장
m.save('1st_day_course.html')  # HTML 파일로 저장
```
